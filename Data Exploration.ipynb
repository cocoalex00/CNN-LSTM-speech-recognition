```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

# Librosa Plot

```python
file = "C:\\_University\\COM3025\\TF-Speech-Recognition-Challenge-Solution\\data\\raw\\train\\audio\\silence\\0a7c2a8d_nohash_1.wav"

signal, sr = librosa.load(file)
librosa.display.waveplot(signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

# Fourier Transform

```python
fft = np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_magnitude = magnitude[:int(len(magnitude)/2)]
left_frequency = frequency[:int(len(frequency)/2)]

plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequence")
plt.ylabel("Magnitude")
plt.show()
```

# Short-time Fourier Transform

```python
n_fft = 2048
hop_length = 512

stft = librosa.core.stft(signal, n_fft=n_fft, hop_length=hop_length)
spectrogram = np.abs(stft)

librosa.display.specshow(spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

# MFCC

```python
mfcc = librosa.feature.mfcc(signal, n_fft=n_fft, hop_length=hop_length, n_mfcc=13)

librosa.display.specshow(mfcc)
plt.xlabel("Time")
plt.ylabel("MFCC")
plt.colorbar()
plt.tight_layout()
plt.show()
```

# MEL

```python
# hop_length = 256
mel = librosa.feature.melspectrogram(signal, n_fft=n_fft, hop_length=hop_length, sr=sr, n_mels=32, fmax=8000)

librosa.display.specshow(librosa.power_to_db(mel, ref=np.max))
# plt.xlabel("Hz")
# plt.ylabel("Mel")
# plt.colorbar()
plt.box(False)
# plt.show()

plt.savefig('C:\\_University\\test.png')
```
